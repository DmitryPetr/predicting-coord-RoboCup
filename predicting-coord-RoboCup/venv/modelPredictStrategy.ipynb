```
#from tensorflow.keras.wrappers.scikit_learn import KerasClassifier
from scikeras.wrappers import KerasClassifier

def build_model(embedding_vector_length=32, load=None):
  model = keras.models.Sequential()
  model.add(
      keras.layers.Embedding(
          46, embedding_vector_length, input_length=max_review_length
      )
  )
  model.add(keras.layers.LSTM(20))
  model.add(keras.layers.Dense(1, activation='sigmoid'))
  if load is not None:
    model.load_weights(load)
    model.trainable = False
  model.compile(
      loss='binary_crossentropy', optimizer='adam', metrics=['accuracy']
  )
  return model

clf = KerasClassifier(build_model)
```

```
#print(np.unique(np.isnan(y_test)))
#print(type(X_train))
#print(type(y_train))
clf.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=20, batch_size=10)
```

```
clf.score(X_test, y_test)
```
